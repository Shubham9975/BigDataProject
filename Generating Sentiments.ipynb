```python
# Function to get sentiment
def get_sentiment(text):
    blob = TextBlob(str(text))  # ensure it's a string
    polarity = blob.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity < 0:
        return 'negative'
    else:
        return 'neutral'
```

```python
# Apply the function to create a new column
revi_df['sentiment'] = revi_df['review_body'].apply(get_sentiment)
```
